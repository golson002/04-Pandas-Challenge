```python
# Import dependencies
import pandas as pd
```

```python
# Define a path to the csv files we want to read
school_data = 'Resources/schools_complete.csv'
student_data = 'Resources/students_complete.csv'
```

```python
# Read the school and student csv files and store their contents in individual Pandas DataFrames
school_df = pd.read_csv(school_data)
student_df = pd.read_csv(student_data)

# Merge the two files into a single data set to work with 
merged_df = pd.merge(school_df, student_df, how = 'left', on = ['school_name','school_name'])
merged_df.head()
```

## District Summary

```python
# Calculate the total schools in the school district using the len function on the merged data set 
total_schools = len(merged_df['school_name'].unique())
total_schools
```

```python
# Calculate the total students in the school district using the len function on the merged data set
total_students = len(merged_df['student_name'])
total_students
```

```python
# Calculate the total budget for the schools in the district
total_budget = sum(merged_df['budget'].unique())
total_budget
```

```python
# Calculate the average math score of students in the district 
average_math = merged_df['math_score'].mean()
average_math
```

```python
# Calculate the average reading score of students in the district 
average_reading = merged_df['reading_score'].mean()
average_reading
```

```python
# Find the number of students in the district who passed math (score >=70) using the len function and .loc on the merged data set
passed_math = len(merged_df.loc[merged_df['math_score'] >= 70])

# Calculate the percentage of students who passed math
passed_math_percent = (passed_math / total_students)*100
passed_math_percent
```

```python
# Find the number of students in the district who passed reading (score >=70) using the len function and .loc on the merged data set
passed_reading = len(merged_df.loc[merged_df['reading_score'] >= 70])

# Calculate the percentage of students who passed reading
passed_reading_percent = (passed_reading / total_students)*100
passed_reading_percent
```

```python
# Find the number of students in the district who passed both math and reading (score >=70) using the len function and .loc on the merged data set
passed_overall = len(merged_df.loc[(merged_df['math_score'] >=70) & (merged_df['reading_score']>=70)])

# Calculate the percentage of students who passed overall
passed_overall_percent = (passed_overall / total_students)*100
passed_overall_percent
```

```python
# Create a District Summary Pandas DataFrame with the district statistics above
district_df = pd.DataFrame({'Total Schools': [total_schools], 
                            'Total Students': total_students,
                            'Total Budget': total_budget,
                            'Average Math Score': average_math,
                            'Average Reading Score': average_reading,
                            '% Passing Math': passed_math_percent,
                            '% Passing Reading': passed_reading_percent,
                            '% Passing Overall': passed_overall_percent})
district_df
```

```python
# Round average scores and % of students passing to two decimal places
district_df['Average Math Score'] = round(district_df['Average Math Score'],2)
district_df['Average Reading Score'] = round(district_df['Average Reading Score'],2)
district_df['% Passing Math'] = round(district_df['% Passing Math'],2)
district_df['% Passing Reading'] = round(district_df['% Passing Reading'],2)
district_df['% Passing Math'] = round(district_df['% Passing Math'],2)
district_df['% Passing Overall'] = round(district_df['% Passing Overall'],2)

# Format 'Total Students' and 'Total Budget' to include commas and dollar sign
district_df['Total Students'] = district_df['Total Students'].map('{:,}'.format)
district_df['Total Budget'] = district_df['Total Budget'].map('${:,.2f}'.format)
district_df
```

## School Summary

```python
# Create a groupby object based on 'school_name'
grouped_schools = merged_df.groupby(['school_name'])
```

```python
# Find the 'Type' of each school using the gschool_df above and setting the index to the 'school_name'
school_type_grouped = school_df.set_index(['school_name'])['type']
school_type_grouped
```

```python
# Find the total number of students at each school using the grouped_schools groupby object
total_students_grouped = grouped_schools['student_name'].count()
total_students_grouped
```

```python
# Find the total budget for each school using the grouped_schools groupby object
total_budget_grouped = grouped_schools['budget'].unique().astype(int)
total_budget_grouped
```

```python
# Find the total budget per student at each school using the total_budget_grouped and total_students_grouped DataSeries 
student_budget_grouped = total_budget_grouped / total_students_grouped
student_budget_grouped
```

```python
# Find the average math score of students at each school using the grouped_schools groupby object
average_math_grouped = grouped_schools['math_score'].mean()
average_math_grouped
```

```python
# Find the average reading score of students at each school using the grouped_schools groupby object
average_reading_grouped = grouped_schools['reading_score'].mean()
average_reading_grouped
```

```python
# Find the names of all students (in the entire district- regardless of school) who passed math
passed_math_names = merged_df.loc[merged_df['math_score'] >= 70]
# Group those names by school
passed_math_grouped = passed_math_names.groupby(['school_name']).count()['student_name']

# Calculate the percentage of students who passed math at each school using the passed_math_grouped groupby object
passed_math_grouped_percent = (passed_math_grouped/total_students_grouped)*100
passed_math_grouped_percent
```

```python
# Find the names of all students (in the entire district- regardless of school) who passed reading
passed_reading_names = merged_df.loc[merged_df['reading_score'] >= 70]
# Group those names by school
passed_reading_grouped = passed_reading_names.groupby(['school_name']).count()['student_name']

# Calculate the percentage of students who passed reading at each school using the passed_reading_grouped groupby object
passed_reading_grouped_percent = (passed_reading_grouped/total_students_grouped)*100
passed_reading_grouped_percent
```

```python
# Find the names of all students (in the entire district- regardless of school) who passed both math and reading
passed_overall_names = merged_df.loc[(merged_df['math_score'] >=70) & (merged_df['reading_score']>=70)]
# Group those names by school
passed_overall_grouped = passed_overall_names.groupby(['school_name']).count()['student_name']

# Calculate the percentage of students who passed overall at each school iusing the passed_overall_grouped groupby object
passed_overall_grouped_percent = (passed_overall_grouped/total_students_grouped)*100
passed_overall_grouped_percent
```

```python
# Create a School Summary Pandas DataFrame with each school's statistics above
grouped_schools_df = pd.DataFrame({'School Type': school_type_grouped,
                        'Total Students': total_students_grouped,
                        'Total School Budget': total_budget_grouped,
                        'Per Student Budget': student_budget_grouped,
                        'Average Math Score': average_math_grouped,
                        'Average Reading Score': average_reading_grouped,
                        '% Passing Math': passed_math_grouped_percent,
                        '% Passing Reading': passed_reading_grouped_percent,
                        '% Passing Overall': passed_overall_grouped_percent})
grouped_schools_df
```

```python
# Round average scores and % of students passing to two decimal places
grouped_schools_df['Average Math Score'] = round(grouped_schools_df['Average Math Score'],2)
grouped_schools_df['Average Reading Score'] = round(grouped_schools_df['Average Reading Score'],2)
grouped_schools_df['% Passing Math'] = round(grouped_schools_df['% Passing Math'],2)
grouped_schools_df['% Passing Reading'] = round(grouped_schools_df['% Passing Reading'],2)
grouped_schools_df['% Passing Overall'] = round(grouped_schools_df['% Passing Overall'],2)

# Format 'Total Students', 'Total School Budget' and 'Per Student Budget' to include commas and currency symbol
grouped_schools_df['Total Students'] = grouped_schools_df['Total Students'].map('{:,}'.format)
grouped_schools_df['Total School Budget'] = grouped_schools_df['Total School Budget'].map('${:,.2f}'.format)
grouped_schools_df['Per Student Budget'] = grouped_schools_df['Per Student Budget'].map('${:,.2f}'.format)
grouped_schools_df
```

## Scores by School Spending

```python
# Determine min and max of each school's per student budget
print(student_budget_grouped.min())
print(student_budget_grouped.max())
```

```python
# Create bins and labels to place the values within the student_budget_grouped series in
bins = [0, 585, 630, 645, 680]
labels = ['<$585','$585-$630','$630-$645','$645-$680']

# Slice the student_budget_grouped data for each school and put it into the correct bin
pd.cut(student_budget_grouped, bins, labels = labels)

# Add the sliced data series into the grouped_schools_df
grouped_schools_df['Per Student Budget Group'] = pd.cut(student_budget_grouped, bins, labels = labels)
grouped_schools_df
```

```python
# Create a Groupby object based on 'Per Student Budget Group'
spending_grouped = grouped_schools_df.groupby('Per Student Budget Group')

# Create a DataFrame to display school performance based on the 'Per Student Budget Group' bins/ranges
spending_grouped_df = spending_grouped[['Average Math Score','Average Reading Score','% Passing Math',
                  '% Passing Reading','% Passing Overall']].mean()
```

```python
# Round average scores and % of students passing to two decimal places
spending_grouped_df['Average Math Score'] = round(spending_grouped_df['Average Math Score'],2)
spending_grouped_df['Average Reading Score'] = round(spending_grouped_df['Average Reading Score'],2)
spending_grouped_df['% Passing Math'] = round(spending_grouped_df['% Passing Math'],2)
spending_grouped_df['% Passing Reading'] = round(spending_grouped_df['% Passing Reading'],2)
spending_grouped_df['% Passing Overall'] = round(spending_grouped_df['% Passing Overall'],2)
spending_grouped_df
```
